```python
#importing the required libraries
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt


#Loading the dataset
df = pd.read_csv('/content/Daily_Public_Transport_Passenger_Journeys_by_Service_Type_20250526.csv')
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df = df.sort_values('Date')

#Removing the 'other' column
if 'Other' in df.columns:
    df = df.drop(columns=['Other'])

#function to forecast
def forecast_service(service_name):
    data = df[['Date', service_name]].rename(columns={'Date': 'ds', service_name: 'y'})

    model = Prophet(daily_seasonality=True, yearly_seasonality=True)
    model.fit(data)

    future = model.make_future_dataframe(periods=7)
    forecast = model.predict(future)
    forecast['yhat'] = forecast['yhat'].clip(lower=0)
    return forecast[['ds', 'yhat']].tail(7)

#Services to forecast
services = ['Local Route', 'Light Rail', 'Peak Service', 'Rapid Route', 'School']
results = {}

for service in services:
    print(f"\nForecast for {service}:")
    forecast_df = forecast_service(service)
    print(forecast_df)
    results[service] = forecast_df

# Plotting the 7-day forecasts for each service
plt.figure(figsize=(12, 6))

for service in services:
    forecast_df = results[service]
    plt.plot(forecast_df['ds'], forecast_df['yhat'], label=service, marker='o')

plt.title('7-Day Forecast of Passenger Journeys by Service Type')
plt.xlabel('Date')
plt.ylabel('Forecasted Passengers')
plt.grid(True)
plt.legend()
plt.show()
```
